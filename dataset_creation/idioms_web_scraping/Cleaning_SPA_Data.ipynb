```python
df2 = pd.read_csv('Spanish_idioms_2.csv')
```

```python
df2.info()
```

```python
df2.head()
```

```python
df2.rename(columns = {'Strong Text':'Idiom','Meaning 1':'Meaning'}, inplace = True) 
df2
```

```python
#check which value have empty meaning 
df2[df2['Meaning'].isnull()]
```

```python
#manually fill the meaning
df2.loc[117, 'Meaning'] = 'Nadie es profeta en su tierra [pueblo]'
```

```python
#remove null values if it has null in idiom column or meaning column
df2.dropna(subset=['Idiom', 'Meaning'], inplace=True)
```

```python
#Remove value with same idiom and meaning
df2.drop_duplicates(subset=['Idiom', 'Meaning'], inplace=True)
```

```python
#Value with not na in meaning 2 create new rows for each meaning 2 - 8 with same idiom
expanded_data = []
for index, row in df2.iterrows():
    base_idiom = row['Idiom']
    expanded_data.append({ 'Idiom': base_idiom, 'Meaning': row.get('Meaning', None) })

    # Check and add additional meanings as separate rows
    for i in range(2, 9):  # Assuming there are up to 8 meanings
        meaning_key = f'Meaning {i}'
        if meaning_key in row and pd.notna(row[meaning_key]):
            expanded_data.append({ 'Idiom': base_idiom, 'Meaning': row[meaning_key] })

# Convert the expanded data back into a DataFrame
df2 = pd.DataFrame(expanded_data)
df2.drop_duplicates(subset=['Idiom', 'Meaning'], inplace=True)
```

```python
df2.info()
```

```python
for column in df2.columns:
    if df2[column].dtype == 'object':  # Check if column is of type object (text)
        df2[column] = df2[column].str.strip()
```

```python
Phrase_counts = df2['Idiom'].value_counts()
Phrase_counts
```

```python
Meaning_counts = df2['Meaning'].value_counts()
Meaning_counts
```

```python
meaning_counts = df2['Meaning'].value_counts()

# Find the idioms that occur more than once
meaning_more_than_once = meaning_counts[meaning_counts> 1].index

# Filter df2 to show only the rows where the 'Idiom' occurs more than once
filtered_df2 = df2[df2['Meaning'].isin(meaning_more_than_once)]

filtered_df2
```

```python
df2.isnull().sum()
```

```python
df2.isna().sum()
```

```python
df2.to_csv('Spanish_idioms_2.csv', index=False)
```

```python
for column in df3.columns:
    if df3[column].dtype == 'object':  # Check if column is of type object (text)
        df3[column] = df3[column].str.strip()
```

```python
Phrase_counts = df3['Idiom'].value_counts()
Phrase_counts
```

```python
Phrase_counts = df3['Idiom'].value_counts()

# Find the idioms that occur more than once
idioms_more_than_once = Phrase_counts[Phrase_counts > 1].index

# Filter df3 to show only the rows where the 'Idiom' occurs more than once
filtered_df3 = df3[df3['Idiom'].isin(idioms_more_than_once)]

filtered_df3
```

```python
# Drop rows with the same meanings and phrases
df3 = df3.drop_duplicates(subset=['Meaning', 'Idiom'])
Phrase_counts = df3['Idiom'].value_counts()

# Find the idioms that occur more than once
idioms_more_than_once = Phrase_counts[Phrase_counts > 1].index

# Filter df3 to show only the rows where the 'Idiom' occurs more than once
filtered_df3 = df3[df3['Idiom'].isin(idioms_more_than_once)]

filtered_df3
```

```python
Meaning_counts = df3['Meaning'].value_counts()
Meaning_counts
```

```python
meaning_counts = df3['Meaning'].value_counts()

# Find the idioms that occur more than once
meaning_more_than_once = meaning_counts[meaning_counts> 1].index

# Filter df3 to show only the rows where the 'Idiom' occurs more than once
filtered_df3 = df3[df3['Meaning'].isin(meaning_more_than_once)]

filtered_df3
```

```python
#Combine all the dataframes
dataframes = [df1, df2, df3]

# Ensure each dataframe has a 'part_of_speech' column, fill with empty strings if it doesn't exist
for df in dataframes:
    if 'literal_translation' not in df.columns:
        df['literal_translation'] = ' '
    if 'Example' not in df.columns:
        df['Example'] = ' '

# Now concatenate the dataframes
result_df = pd.concat(dataframes, axis=0, join='outer', ignore_index=True)

# If you only want 'Idiom', 'Example', 'Meaning', and 'part_of_speech' columns in the result
result_df = result_df[['Idiom', 'Meaning','literal_translation','Example']]

result_df
result_df.to_csv('Spanish_idioms.csv', index=False)
```
